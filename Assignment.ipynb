```python
# Using random forest ( since it is a ensemble technique and use decision tree , therefore no scaling is required)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
df.corr() #feature 6 has highest correlation (0.26)
```

```python
rf=RandomForestClassifier(random_state=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
param_grid= { 'max_features':['auto','log2','sqrt'] , 'max_depth':[3,4,5,6,7] }
```

```python
model_temp=GridSearchCV(estimator=rf,param_grid=param_grid,cv=5)
```

```python
model_temp.fit(X_train,y_train)
```

```python
model_temp.best_params_
```

```python
pred=model_temp.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
fa=accuracy_score(y_test,pred)
```
